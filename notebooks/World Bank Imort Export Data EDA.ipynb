# Load csvs

```python
import pandas as pd
```

## Imports

```python
imports = pd.read_csv('../raw_data/imports_wb.csv', skiprows=4)
```

```python
imports
```

```python
imports.drop(columns=['Country Name', 'Unnamed: 67', 'Indicator Code', 'Indicator Name'], inplace=True)
```

```python
imports
```

```python
imports = pd.DataFrame(imports.set_index('Country Code').stack()).reset_index().rename(columns={'level_1': 'Year', 0: 'Imports'})
```

```python
imports
```

```python
mapper = pd.read_csv('../data/countries_info.csv').loc[:, ['Alpha-2 code', 'Alpha-3 code']]
```

```python
df = imports.merge(right= mapper, left_on='Country Code', right_on='Alpha-3 code').drop(columns=['Country Code', 'Alpha-3 code']).rename(columns={'Alpha-2 code': 'Destination country', 'Destination country': 'Country'})
```

```python
df
```

```python
df.loc[:, 'source_hash'] = 'ae91c06b2fac22ef2ed0e9fecb2368245e17b8dc843db2d211a07d0e893e5238'
```

```python
df.to_csv('../data/imports.csv')
```
