## Import Data per country 

```python
imports_df = df[df['Category'] == ' c']
```

```python
imports_df[imports_df['Destination country'] == 'MM']
```

```python
imports_df = imports_df.groupby(by = ['Destination country', 'Year']).sum()
```

```python
imports_df.reset_index(inplace=True)
```

```python
imports_df
```

```python
imports_df.loc[:, 'Destination country'] = imports_df.loc[:, 'Destination country'].str.strip()
```

```python
imports_df.drop(columns=['Embargo'], inplace=True)
```

```python
imports_df.to_csv('data/imports.csv')
```
