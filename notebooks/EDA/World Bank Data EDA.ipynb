# Load csv files

```python
import pandas as pd
```

## Armed Forces Personnel (Total)

```python
armed_forces_personnel = pd.read_csv('../raw_data/world_bank_data/armed_forces_personnel.csv', skiprows=4)
```

```python
armed_forces_personnel.drop(columns=['Country Name', 'Unnamed: 67', 'Indicator Code', 'Indicator Name'], inplace=True)
armed_forces_personnel = pd.DataFrame(armed_forces_personnel.set_index('Country Code').stack()).reset_index().rename(columns={'level_1': 'Year', 0: 'Total Armed Forces Personel'})
```

```python
armed_forces_personnel
```

# Merge stuff


## Merge all datadrames


```python
merged_df = milit_exp_gdp.merge(right=milit_exp_gov, how='outer').merge(right=milit_exp_usd, how='outer').merge(right=armed_forces_personnel, how='outer').merge(right=armed_forces_personnel_per, how='outer')
```

```python
merged_df
```

## Megre in country code

```python
mapper = pd.read_csv('../data/countries_info.csv').loc[:, ['Alpha-2 code', 'Alpha-3 code']]
```

```python
df = merged_df.merge(right= mapper, left_on='Country Code', right_on='Alpha-3 code').drop(columns=['Country Code', 'Alpha-3 code']).rename(columns={'Alpha-2 code': 'Destination country', 'Destination country': 'Country'})
```
